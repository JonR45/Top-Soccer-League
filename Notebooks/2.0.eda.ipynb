# Exploratory Data Analysis

## Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load data

```python
raw_data = pd.read_csv('../Data/Final/players_outfield_prepared.csv')
data = raw_data.copy()
df = pd.DataFrame(data)
pd.set_option('display.max_columns', None)
data.head()
```

I want to understand if there are significant differences between stats for players in the top 5 leagues vs. players who aren't. I'll do this by exploring the means and medians.

## Group by Top 5 league

```python
df_mean = df.groupby(['Top_5_league']).mean()
df_mean
```

```python
df_median = df.groupby(['Top_5_league']).median()
df_median
```

```python
df_mode = df.groupby(['Top_5_league']).agg(pd.Series.mode)
df_mode
```

- From analysing the above I can eliminate many features where there isn't a clear difference between the top 5 league players and their lower division counterparts. 
- I'm then left with features that appear to be more useful.
